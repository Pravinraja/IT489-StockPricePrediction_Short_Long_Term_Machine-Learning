```python
import os
import tweepy
import pandas as pd

# use os.environ.get to obtain other environment variables
# from ~/.bashrc or ~/.zshrc etc., so they aren't in your code
consumer_key = ''
consumer_secret = ''
access_token = ''
access_token_secret = ''
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)

results = api.search(q='Tesla Stock', count=100)
results1 = api.search(q='Pfizer Stock', count=100)
results2 = api.search(q='Apple Stock', count=100)
results3 = api.search(q='Microsoft Stock', count=100)
results4 = api.search(q='Google Stock', count=100)

json_data = [r._json for r in results]
json_data1 = [r._json for r in results1]
json_data2 = [r._json for r in results2]
json_data3 = [r._json for r in results3]
json_data4 = [r._json for r in results4]


df = pd.io.json.json_normalize(json_data)
df_PFE = pd.io.json.json_normalize(json_data)
df_AAPL = pd.io.json.json_normalize(json_data)
df_MSFT = pd.io.json.json_normalize(json_data)
df_GOOG = pd.io.json.json_normalize(json_data)
```
